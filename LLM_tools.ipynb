```python
from tavily import TavilyClient

# Create client only once
tavily_client = TavilyClient()
```


Web Search


```python
def web_search(query: str, max_results: int = 5) -> dict:
    """
    Perform a normal web search using Tavily
    """
    result = tavily_client.search(
        query=query,
        max_results=max_results
    )
    return result
```

```python
response = web_search("What is Feynman technique?")
print(response)
```

News Search

```python
def news_search(query:str , days:int = 7) ->dict:
    "Search recent news using Tavity"
    result = tavily_client.search(
        query=query,
        topic="news",
        days=days
    )
    return result
```

```python
news = news_search("what is the latest news about Brain-computer interface(BCI) chips")
print(news)
```

Question-Answer


```python
def qa_search(question: str) -> dict:
    """
    Direct answering usin tavity
    """
    result = tavily_client.search(
        query=question,
        include_answer = True,
    )
    return result
```

```python
qa = qa_search("Who is the founder of R1 robot neuralink?")
print("Answer: ",qa["answer"])
```

Image Search

```python
def image_search(query: str) -> dict:
    """
    Image related content using tavily
    """
    result = tavily_client.search(
        query = query,
        include_image = True
    )

    return result
```

```python
image = image_search("Taj mahal")
print(image["images"])
```

```python
result = image_search("R1 robot neuralink")
print(result.keys())
```

```python
print(result["results"])
```

Summary search


```python
def summary_search(query: str) -> dict:
    """
    Combined summary from multiple
    """

    result = tavily_client.search(
        query=query,

    )

    summary = ""
    for item in result["results"]:
        summary += item["content"][:200] + " "

        return summary
```

```python
print(summary_search("Is gut microbes affects mental health?"))
```

Region based search


```python
def region_based_search(topic: str, region: str) -> dict:
    """
    Search topic with regional context
    """
    query = f"{topic} in {region}"
    result = tavily_client.search(
        query=query,
        max_results=5
    )
    return result
```

```python
region_based_search("Iron rich foods for women", "rural Maharashtra")
```
