
Web Search


```python
def web_search(query: str) -> dict:
    """ Perform a normal web search using Tavily """
    result = tavily_client.search(
        query=query
    )
    return result
```

```python
response = web_search("What is Feynman technique?")
print(response)
```

News Search

```python
def news_search(query:str ) ->str:
    "Search recent news using Tavity"
    result = tavily_client.search(
        query=query,
        topic="news"
    )
    return result
```

```python
news = news_search("what is the latest news about Brain-computer interface(BCI) chips")
print(news)
```

Question-Answer


```python
def qa_search(question: str) -> dict:
    """ Direct answering usin tavity """
    result = tavily_client.search(
        query=question,
        include_answer = True,
    )
    
    return result
```

```python
qa = qa_search("Who is the founder of R1 robot neuralink?")
print("Answer: ",qa["answer"])
```

Image Search

```python
def image_search(query: str) -> dict:
    """ Image related content using tavily """
    result = tavily_client.search(
        query = query,
        include_image = True
    )

    return result
```

```python
image = image_search("Taj mahal")
print(image["images"])
```

```python
result = image_search("R1 robot neuralink")
print(result.keys())
```

Summary search


```python
def summary_search(query: str) -> dict:
    """ Combined summary from multiple """

    result = tavily_client.search(
        query=query,

    )

    summary = ""
    for item in result["results"]:
        summary += item["content"][:200] + " "

        return summary
```

```python
print(summary_search("Is gut microbes affects mental health?"))
```

Region based search


```python
def region_based_search(topic: str, region: str) -> dict:
    """
    Search topic with regional context
    """
    query = f"{topic} in {region}"
    result = tavily_client.search(
        query=query,
        max_results=5
    )
    return result
```

```python
region_based_search("Iron rich foods for women", "rural Maharashtra")
```

```python
from langchain.tools import tool
import wikipedia

@tool("wikipedia_summary")
def wikipedia_summary(query: str) -> str:
    """
    Get a short factual summary from Wikipedia for the given query.
    """
    return wikipedia.summary(query, sentences=7)
```

```python
result = wikipedia_summary.invoke("Feynman technique")
print(result)
```

Calculator


```python
@tool ("calculator")
def calculator(query:str) ->str:
    """ Mathematical calculations"""
    return str(eval(query))
```

```python
result = calculator.invoke("2+5+4")
print(result)
```

```python
from langgraph.types import Command
from langchain.messages import RemoveMessage
from langgraph.graph.message import REMOVE_ALL_MESSAGES
from langchain.tools import tool, ToolRuntime

@tool 
def clear_conversation() -> Command:
    """ Clear the conversation history"""
    return Command(
        update={
            "messages" : [RemoveMessage(id=REMOVE_ALL_MESSAGES)] ,
        }
    )
```

```python
result = clear_conversation.invoke({})
print(result)
```

Helper Funtions

```python
def format_result(tavily_response):
    """
    Extract title, content, and url from Tavily response
    """
    formatted = []

    for item in tavily_response.get("results", []):
        formatted.append({
            "title": item.get("title"),
            "content": item.get("content"),
            "url": item.get("url")
        })

    return formatted
```

```python
formatted_results = format_result(response)
print(formatted_results)
```

```python
def summarise_web(content):
    """
    Create a short summary from web content
    """
    if not content:
        return ""
    return content[:200] + "..."
```

```python
def deduplication(results):
    """
    Remove duplicate search results based on URL
    """
    seen_urls = set()
    unique_results = []

    for item in results:
        url = item.get("url")
        if url and url not in seen_urls:
            seen_urls.add(url)
            unique_results.append(item)

    return unique_results
```

```python
unique_results = deduplication(formatted_results)
print(unique_results)
```

```python
summary = summarise_web(unique_results[0]["content"])
print(summary)
```
