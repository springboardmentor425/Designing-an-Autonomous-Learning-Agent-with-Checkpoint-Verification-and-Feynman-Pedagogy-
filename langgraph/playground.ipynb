```python
from langgraph.graph import START,END,StateGraph
from typing import TypedDict,Literal
```

```python
#state
class State(TypedDict):
  operation:str
  operand_a:float
  operand_b:float
  result:float
```

```python
#add operation
def node_add(State):
  a=State["operand_a"]
  b=State["operand_b"]
  result=a+b
  return {"result":result}
```

```python
#subtraction operation
def node_subtract(State):
  a=State["operand_a"]
  b=State["operand_b"]
  result=a-b
  return {"result":result}
```

```python
#multiplication operation 
def node_multiply(State):
  a=State["operand_a"]
  b=State["operand_b"]
  result=a*b
  return {"result":result}
```

```python
#division operaation
def node_divide(State):
  a=State["operand_a"]
  b=State["operand_b"]
  result = a*b
  return {"result":result}
```

```python
#check operation
def check_operation(State)-> Literal["node_add","node_subtract","node_multiply","node_divide"]:
  if State["operation"]=="add":
    return "node_add"
  elif State["operation"]=="subtract":
    return "node_subtract"
  elif State["operation"]=="multiply":
    return "node_multiply"
  else :
    return "node_divide"
```

```python
builder=StateGraph(State)
builder.add_node("node_add",node_add)
builder.add_node("node_subtract",node_subtract)
builder.add_node("node_multiply",node_multiply)
builder.add_node("node_divide",node_divide)

builder.add_conditional_edges(START,check_operation)
builder.add_edge("node_add",END)
builder.add_edge("node_subtract",END)
builder.add_edge("node_multiply",END)
builder.add_edge("node_divide",END)
```

```python
graph=builder.compile()

result=graph.invoke({"operation":"subtract","operand_a":8,"operand_b":4})
print(result)
```
